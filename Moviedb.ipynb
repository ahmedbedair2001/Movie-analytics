<h2>Importing libraries</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
sns.set()
```

<h2>Reading and exploring data</h2>

```python
md=pd.read_csv('mymoviedb.csv',engine='python')
```

```python
md.info()
```

```python
md.describe()
```

<h2>Dropping na values </h2>

```python
md.dropna(inplace=True,subset=['Title'])
```

```python
md
```

```python
md.drop(labels=['Poster_Url','Overview'],axis=1,inplace=True)
```

```python
md['Vote_Average']=pd.to_numeric(md['Vote_Average'], errors='coerce')
md['Vote_Count']=pd.to_numeric(md['Vote_Count'], errors='coerce')
md['Release_Date']=pd.to_datetime(md['Release_Date'], errors='coerce')
```

```python
md.info()
```

```python
md[md['Genre'].isna()]
```

```python
md.dropna(inplace=True,subset=['Genre'])
```

```python
md.info()
```

<h2>Using title as index</h2>

```python
md.set_index('Title',inplace=True)
```

```python
md
```

<h2>Q1: What is the most popular type of movie?</h2>

```python
plt.barh(md['Genre'].value_counts().head().index,md['Genre'].value_counts().head())
plt.show()
```

```python
print('Most popular type is '+md['Genre'].value_counts().index[0]+" By "+str(md['Genre'].value_counts()[0])+" movie")
```

<h2>Q2: What kind of movies have the highest rating?</h2>

```python
rm=md.groupby(md['Genre']).mean()
```

```python
rm
```

```python
rm.sort_values(by=['Vote_Average'],inplace=True,ascending=False)
```

```python
rm
```

```python
plt.barh(rm['Vote_Average'].head().index,rm['Vote_Average'].head())
plt.show()
```

```python
print('Most rated kind is ('+rm['Vote_Average'].index[0]+") By rate: "+str(rm['Vote_Average'][0]))
```

<h2>Q3: What is the most watched movie?</h2>

```python
md.sort_values(by=['Popularity'],inplace=True,ascending=False)
```

```python
plt.barh(md['Popularity'].head().index,md['Popularity'].head())
plt.show()
```

```python
print('Most watched movie is ('+md['Vote_Average'].index[0]+')')
```

<h2>Q4: What is the year that most films have been produced?</h2>

```python
md.sort_values(by=['Release_Date'],inplace=True,ascending=False)
```

```python
md
```

```python
reyear=md['Release_Date'].dt.year
reyearvc=reyear.value_counts().head()
reyearvc
```

```python
reyearvc.plot(kind='bar',figsize=(15,10))
plt.show()
```

```python
print(str(reyearvc.index[0])+' is the year that most films have been produced')
```
